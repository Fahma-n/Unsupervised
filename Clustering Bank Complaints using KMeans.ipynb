```python

dt4["Sub-product"].fillna("No", inplace = True)



dt4["State"].fillna("No", inplace = True)
dt4["ZIP code"].fillna("No", inplace = True)


dt4["Consumer disputed?"].fillna("No", inplace = True)



```

```python

dt4.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
le=LabelEncoder()
w=('Date received', 'Product', 'Sub-product', 'Issue', 
         'Company',
       'State', 'ZIP code', 
       'Submitted via', 'Date sent to company', 'Company response to consumer',
       'Timely response?', 'Consumer disputed?', 'Complaint ID')
for i in w:
   dt4[i]=le.fit_transform(dt4[i])
data1=dt4.head(21)
data1
```

```python
import seaborn as sns
sns.barplot(x="Complaint ID", y="Issue", data=data1)
```

```python
sns.jointplot(x="Complaint ID", y="Issue", data=data1)
```

```python
sns.heatmap(data1.corr(),annot=True,cmap="Blues")
```

```python
from scipy.spatial.distance import cdist
from sklearn.cluster import KMeans
datas=[]
for i in range(1,11):
    km=KMeans(n_clusters=i)
    km.fit(data1)
    datas.append(sum(np.min(cdist(data1,km.cluster_centers_,"euclidean"),axis=1))/data1.shape[0])
print(datas)
```

```python
import matplotlib.pyplot as plt
k=range(1,11)
print(list(k))
plt.plot(k,datas,"bx-")
plt.xlabel("k")
plt.ylabel("datas")
plt.title("select k value")
plt.show()
```

```python
km1=KMeans(n_clusters=5)
km1.fit(data1)
```

```python
pr=km1.predict(data1)
pr
```

```python
hello=data1['Issue']
#hello.values.reshape(len(hello),1)
yes=data1['Product']
#yes.values.reshape(len(yes),1)
plt.scatter(hello,yes,c=pr)
centroids=km1.cluster_centers_
centroids_x=centroids[:,0]
centroids_y=centroids[:,1]
color=["red","blue","orange","pink","black"]
plt.scatter(centroids_x,centroids_y,marker="*",s=50,color=color)
plt.show()

#centroids=km1.cluster_centers_
#centroids_x=centroids[:,0]
#centroids_y=centroids[:,2]
#color=["red","blue"]
#plt.scatter(km1.cluster_centers_[:,0],km1.cluster_centers_[:,1],s=50,c=color,marker="*",label="center")
#plt.scatter(centroids_x,centroids_y,marker="*",s=50,color=color)
#plt.show()
```
